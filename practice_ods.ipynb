```python
cmap_opt=['cividis','PuBuGn','Wistia','Spectral','twilight','gist_stern']
```

```python
im_z=im[:,:,0]
for i in range(1,len(cmap_opt)+1):
  plt.subplot(2,3,i)
  plt.imshow(im_z,cmap=cmap_opt[i-1])
  plt.title(cmap_opt[i-1])
  plt.axis('off')
  plt.tight_layout(pad=2)
```

```python
import matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages
import numpy as np

with PdfPages('colormap_outputs.pdf') as pdf:
  plt.figure(figsize=(20, 20))
  for i in range(1, len(cmap_opt) +1):
        plt.subplot(2,3,i)
        plt.imshow(im_z, cmap=cmap_opt[i-1])
        plt.title(cmap_opt[i-1])
        plt.axis('off')
        plt.tight_layout(pad=2)

        pdf.savefig()
        # Save current figure to PDF
        plt.close()
        # Close the figure to avoid display and memory buildup
```
